```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Biểu đồ đường thẳng
x = [1, 2, 3, 4]
y = [5 ,6 ,7 ,8]
z = [0, 2 ,4, 6]

# plot(x, y): Vẽ đồ thị dựa trên các tập hợp x và y
# Tham số thứ 3 trong plot() là format cách biểu diễn
# Các kí tự chữ cái tượng trưng cho màu sắc
# VD 'o' là dấu tròn, '-' là đường nét liền, '--' là đường nét đứt, '^' là tam giác
plt.plot (x, y, 'ro')
plt.plot(x, z, 'b--')
plt.plot(z,y)

# title(): Đặt tên đồ thị
# xlabel(): Đặt tên trục hoành
# ylabel(): Đặt tên trục tung
# legend(): Mô tả các kí hiệu
plt.title("Basic plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(["x", "z"])

# show(): Hiển thị biểu đồ
plt.show()
```

```python
# Biểu đồ tròn
# Chuẩn bị dữ liệu
data = pd.read_csv('data1.csv')
data
```

```python
# Tính toán phần trăm ngôn ngữ có trong dữ liệu
languages = data["Language"].to_list()
languageList = list(set(languages))

def countPercentage(languages, languageList):
    countList = []
    for language in languageList:
        count = 0
        for lang in languages:
            if lang == language:
                count+=1
        countList.append(count / len(languages) * 100)
    return countList
    
percentage = countPercentage(languages, languageList)
```

```python
# Sử dụng hàm pie() để vẽ biểu đồ tròn
labels = languageList
data = percentage
plt.pie(data,None, labels, autopct='%1.1f%%',startangle=90)
plt.show()
```

```python
# Biểu đồ cột
data = ['Rust', 'Python', 'TypeScript', 'Kotlin', 'WebAssembly', 'Swift', 'Clojure', 'Elixir', 'Go', 'C#', 'JavaScript', 'Dart']
ranking = np.arange(1, len(data) + 1, 1)
data.reverse()

plt.xlabel('Ranking - Higher is better')
plt.title('Most Loved languages on StackOverflow')

# Sử dụng hàm barh() để vẽ biểu đồ cột ngang
# bar để vẽ biểu đồ cột dọc
plt.barh(data, ranking)
plt.plot()
```

```python
# Subplot
# Để vẽ biểu đồ con (subplot) ta cần phải tạo figure
# Sau đó thêm các biểu đồ con vào figure
# hàm subplot nhận đầu vào là 3 chữ số nguyên tượng trưng cho kích cỡ và vị trí của biểu đồ con
a = [1 ,2 ,3 ,4]
b = [2, 4, 6 ,8]
c = [-2, -4, -6, -8]
d = [1, 3, 5, 7]

fig = plt.figure()
subplot1 = plt.subplot(211)
subplot2 = plt.subplot(212)

subplot1.plot(a,b)
subplot2.plot(c,d)

fig.add_subplot(subplot1)
fig.add_subplot(subplot2)

plt.show()
```
